Reading the Training Data

```
train=pd.read_csv("/content/drive/My Drive/train.csv")
```

```
train.head()
```

Adding one more column "new_text" which is a combination of "title" and "text"

```
train["new_text"]=train["title"]+train["text"]
train
```

```
train
# dropping all the null values which are very less and can be dropped as it will not impact the model
train.dropna(inplace=True)
x=train["new_text"]
y=train["label"]
```

Importing stopwords, wordnet , WordNetLemmatizer, pos_tag to remove all the stopwords and lemmatize the text

```
from nltk.corpus import stopwords
from nltk.corpus import wordnet
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()

# This function returns the pos_tag of the words so that it can be lemmatized
def get_simple_pos(tag):
    if(tag.startswith("J")):
     #   print(1)
        return wordnet.ADJ
    elif(tag.startswith("V")):
     #   print(2)
        return wordnet.VERB
    elif(tag.startswith("N")):
     #   print(3)
        return wordnet.NOUN
    elif(tag.startswith("R")):
     #   print(4)
        return wordnet.ADV
    else:
     #   print(5)
        return wordnet.NOUN
```

Lemmatizing all the stopwords and lemmatizing all the words in each and every text and storing in corpus list

```
corpus=[]
for i in range(len(x)):
    x.iloc[i]=re.sub("[^a-zA-Z]"," ",x.iloc[i])
    review=x.iloc[i].split()
    new=[]
    for j in range(len(review)):
        review[j]=review[j].lower()
        if(review[j] not in stopwords.words("english")):
            review[j]=lemmatizer.lemmatize(review[j],pos=get_simple_pos(pos_tag([review[j]])[0][1]))
            new.append(review[j])
    new_str=" ".join(new)
    corpus.append(new_str)
    print(i,"--------->",len(x)-1)
```

```
from tensorflow.keras.preprocessing.text import one_hot
```

```
# Vocabulary size-----> voc_size=50000, each and every word will be replaced to its index number in the 50000 length vocabulary after one hot encoding
voc_size=50000
```

Performing One Hot Encoding

```
one_hot_data=[one_hot(words,voc_size) for words in corpus]
```

```
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Bidirectional
from tensorflow.keras.layers import Dense
```

Padding our vectorized training data so that each and every text will have same length by adding 0's in the begigning

```
# Sent_length------> Dimensio
sent_length=5000
embedded_doc=pad_sequences(one_hot_data,padding="pre",maxlen=sent_length)
```

```
model=Sequential()
```

Embedding our training data so that each and every word in out text will have some dependency to it's following words in the text

```
model.add(Embedding(voc_size,400,input_length=sent_length))
model.add(Bidirectional(LSTM(100)))
model.add(Dense(1,activation="sigmoid"))
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
model.summary()
```

Training our model with training data with 15 epochs and 64 batch size

```
x_train=np.array(embedded_doc)
y_train=np.array(y)
model.fit(x_train,y_train,epochs=15,batch_size=64)
```
